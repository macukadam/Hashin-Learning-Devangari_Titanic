# Naive Bayesian Solution

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
import pandas as pd

trainData=pd.read_csv('train.csv')
testData=pd.read_csv('test.csv')

le = LabelEncoder()

trainHeaders = trainData.columns.values
testHeaders = testData.columns.values


for i in trainHeaders:
    if(i=='Cabin' or i=='Embarked'):
        trainData[i] = le.fit_transform(trainData[i].fillna('0'))
    else:
        trainData[i] = le.fit_transform(trainData[i])

for i in testHeaders:
    if(i=='Cabin' or i=='Embarked'):
        testData[i] = le.fit_transform(testData[i].fillna('0'))
    else:
        testData[i] = le.fit_transform(testData[i])

train_X = trainData.iloc[:,0:12].drop('Survived', axis=1)
train_y = trainData.iloc[:,1:2]

test_X = testData.iloc[:,0:11]

gnb = GaussianNB()

gnb.fit(train_X, train_y)

df = pd.DataFrame(gnb.predict(test_X),columns=['Survived'])

con = pd.concat([test_X, df], axis=1)
con = con[['PassengerId','Survived']]
con['PassengerId'] += 892

con.to_csv('csvexample.csv', index=False)

test_X
```
